```python
#EMA function to calulate the Exp Moving Avg
def EMA_12(df):
    return df['Close'].ewm(span = 12 , adjust = False , min_periods = 12).mean()
```

```python
df['12 day EMA'] = EMA_12(df)
```

```python
def EMA_26(df):
    return df['Close'].ewm(span =26,adjust = False , min_periods = 26).mean()
```

```python
df['26 day EMA'] = EMA_26(df)
```

```python
df['MACD'] =EMA_12(df) - EMA_26(df)
```

```python
def EMA_9day_MACD(df) :
    return df['MACD'].ewm(span = 9 , adjust = False , min_periods = 9).mean()
```

```python
df['Signal Line'] = EMA_9day_MACD(df)
```

```python
# To Calculate the RSI 

diff = df['Close'].diff(1)
gain = diff.copy()
loss = diff.copy()
gain[gain<0]=0
loss[loss>0]=0
df['Gain']= gain
df['Loss'] = loss
Avg_gain = df['Gain'].rolling(window = 14).mean()
Avg_loss = abs(df['Loss'].rolling(window = 14).mean())
    
RS =  Avg_gain/Avg_loss
RSI =  100 - (100/(1+RS))
df['RSI'] = RSI
```

```python
# Function to calculate SMA
def SMA(df):
    return df['Close'].rolling(window = 20).mean()
```

```python
df['Middle line'] = SMA(df)
```

```python
# Standard Deviation on the 20 day SMA 
stdev = df['Close'].rolling(window = 20).std()
```

```python
df['Standard Dev'] = stdev
```

```python
df['Upper Band'] = (SMA(df) +(2*stdev))
```

```python
df['Lower Band'] = (SMA(df)-(2*stdev))
```

```python
def evening_star(df):
    pattern_detected = []

    for i in range(2, len(df)):
        previous_day = df.iloc[i - 2]
        current_day = df.iloc[i - 1]
        next_day = df.iloc[i]
        
        if previous_day['Close'] > previous_day['Open'] and\
        abs(current_day['Close'] - current_day['Open'])<2 and\
        next_day['Close']< next_day['Open'] and\
        next_day['Open'] > ((previous_day['Close'] + previous_day['Open'])/2)and\
        next_day['Close'] < previous_day['Open']:
            pattern_detected.append((i,df.index[i]))

    return pattern_detected
```

```python
pattern_indices = evening_star(df)

if pattern_indices:
    print("Evening Star pattern detected at indices:", pattern_indices)
else:
    print("No Evening Star pattern detected.")
```

```python
# Strategies

# Using RSI-

# 0-30 is an oversold region and 70-100 is overbought so we assume that the price will correct itself.
# if it reaches 0-30 we must prepare to go long and if it reaches 70-100 we must short the stock
# (except if it is stuck in the region for a long time) then perform the reverse.
# Confirm the idea with S&R and volumes.
# Look for a good pattern to make a more confident decision

# MACD-

# If the MACD shows divergence(turns from positive to negative) we shall consider buying 
# If the MACD shows convergence(turns from negative to positive) we shall consider selling 
# But run into trouble when the MACD moves to zero and turns back(Can be solved by using a crossover of 'signal line' & 'MACD')
# When MACD crosses above the signal line (consider buying) and When the signal line crosses above the MACD line(short trade)
# Useful with a trend

#Bollinger Bands-

# The price expected to stay closer to the average . If it reaches the lower band(buy).If it reaches the upper band(short)
# Works well in a sideways market

#Evening Star-

# Short the stock on P3(Strong red candle day after the spinning top/doji) 
# Confirm if it is at the end of an uptrend
# Stopless is the highest of P1,P2,P3



# Stochastic Oscillator-

# %K crossing below %D - Sell signal and %K above %D - buy signal
# Divergence between the price and the oscillator could indicate a potential reversal
# Check for overbought and oversold(same as RSI)
```

```python
def backtest_strategy(df):
    signals = []
    position = None
    target = None
    stop_loss = None
    for i in range(len(df)):
        data =df.iloc[i]
        if data['RSI'] <= 20 or  data['Close'] <= data['Lower Band']:
            if position is None:
                signals.append('Buy')
                position = 'Long'
                target = data['Close'] * 1.05 
                stop_loss = data['Close'] * 0.90 
            elif position == 'Short':
                signals.append('Buy')
                position = 'Long'
                target = data['Close'] * 1.10  
                stop_loss = data['Close'] * 0.90
            else:
                signals.append('Hold')
                
        elif data['RSI'] >= 80 or  data['Close']>= data['Upper Band']:
            if position is None:
                signals.append('Short')
                position='Short'
                target = data['Close'] * 0.90  
                stop_loss = data['Close'] * 1.10
            elif position == 'Long':
                signals.append('Short')
                position = 'Short'
                target = data['Close'] * 0.90  
                stop_loss = data['Close'] * 1.10
            else:
                signals.append('Hold')
        else:
            signals.append('Hold')
            
    return signals

```

```python
def calculate_metrics(df, initial_capital):
    nav = initial_capital
    pnl_list = []
    position = None
    target = None
    stop_loss = None
    profitable_trades = 0
    total_trades = 0
    total_pnl = 0
    min_loss = float('inf')
    max_profit = float('-inf')
    total_profit = 0
    total_loss = 0
    max_drawdown = 0
    
    for i in range(len(df)):
        data = df.iloc[i]
        pnl = 0
        
        if data['Signals'] == 'Buy' and position is None:
            position = 'Long'
            target = data['Close'] * 1.05
            stop_loss = data['Close'] * 0.95
            total_trades += 1
        elif data['Signals'] == 'Short' and position is None:
            position = 'Short'
            target = data['Close'] * 0.95
            stop_loss = data['Close'] * 1.05
            total_trades += 1
        elif data['Signals'] == 'Hold' and position is not None:
            if position == 'Long' and data['Close'] <= target:
                pnl = target - data['Close']
                profitable_trades += 1
                total_pnl += pnl
                total_profit += pnl
                max_profit = max(max_profit, pnl)
            elif position == 'Short' and data['Close'] >= target:
                pnl = data['Close'] - target
                profitable_trades += 1
                total_pnl += pnl
                total_profit += pnl
                min_loss = min(min_loss, pnl)
                max_profit = max(max_profit, pnl)
            pnl_list.append(pnl)
            nav += pnl
            position = None
            target = None
            stop_loss = None
            
    
    nav = np.cumsum(pnl_list) + initial_capital 
    drawdown = (nav - np.maximum.accumulate(nav)) / np.maximum.accumulate(nav)   
    max_drawdown = drawdown.min() 
    hit_rate = profitable_trades / total_trades if total_trades > 0 else 0
    roi = total_pnl / initial_capital
    avg_profit = total_profit / profitable_trades if profitable_trades > 0 else 0
    avg_loss = total_loss / (total_trades - profitable_trades) if (total_trades - profitable_trades) > 0 else 0
    reward_to_risk_ratio = avg_profit / abs(avg_loss) if avg_loss != 0 else 0
    
    metrics = {
        'Max Drawdown': max_drawdown,
        'Hit Rate': hit_rate,
        'ROI': roi,
        'Minimum Loss': min_loss,
        'Maximum Profit': max_profit,
        'Reward to Risk Ratio': reward_to_risk_ratio
    }
    
    return metrics
```
